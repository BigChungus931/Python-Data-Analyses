# Saving and Loading Numpy Arrays

## Saving Arrays

np.save() Saves your array as a .npy file. <br>
The npy file is like a sealed lunchbox. <br>
It keeps your food (numbers) fresh and exactly how you packed them and it is not meant to be open by other apps.<br>
Just Numpy.

```python
import numpy as np
```

```python
score = np.array([95, 86, 42, 78, 29])

np.save("scores.npy", score)
```

## Loading Arrays

np.load() Opens the array which was saved as a .npy file.

```python
loaded = np.load("scores.npy")

print(loaded)
```

### Text Format - easy to share

np.savetxt() Saves your array as a plain text or csv file. <br>
Something anyone can open in excel, notebook, or Google sheets

```python
temp = np.array([12.1, 24, 32.9, 16, -4])

names = np.array(["Daniel", "Abraham", "Andrew"])

np.savetxt("temp.csv", temp, delimiter = ",", fmt = "%.2f")

np.savetxt("names.csv", names, delimiter = ",", fmt = "%s")
```

**fmt** stands for format. <br>
It tells numpy how to write each number as text. <br>

| fmt | meaning | output |
|-----|---------|--------|
|"%d" | integer | 1, 2, 3|
|"%.2f"| float with 2 decimal places  | 1.00, 2.00, 3.00 |
|"%.4f"| float with 4 decimal places | 1.0000, 2.0000, 3.0000 |
|"%e"  | scientific notation |1.00e+00, 2.00e+00, 3.00e+00 |
|"%s"| string (text) | Hello |
