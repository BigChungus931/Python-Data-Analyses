# Handling Missing and Special Values

## np.nan

Nan stands for not a number. <br>
np.nan is numpy's way of representing a missing numeric value.

```python
import numpy as np
```

```python
data = np.array([[10, np.nan, 30]])

print(np.sum(data))
```

## np.isnan

np.isnan can be understood as an if statement that checks if a slot or a value in an array is set to np.nan. <br>
np.isnan returns as a boolean value (True or False).

```python
print(np.isnan(data))
```
