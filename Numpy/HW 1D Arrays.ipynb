Given an Array with values ([0,2,5,3,7,18,20,21,11,14]). <br>
1. Find the maximum number in this arrays?
2. Find the index for the maximum number in this array?
3. Added the maximum number in this array to the value that follows this number in the array?

```python
arr = np.array([0,2,5,3,43,7,18,20,21,11,14])
print(np.max(arr))
print(np.argmax(arr))
m_index = np.argmax(arr) # m_index equals to the index of the max number which is 43
n_index = m_index + 1 # n_index equals to the m_index + 1 which is 7
sum = arr[m_index] + arr[n_index] # sum equals to addition of both m and n index
print(sum) # print sum which gives 50
```
