# Module 3 Project

Load a CSV of daily temperature readings using np.loadtxt(). Calculate mean, min, max, and standard
deviation for each month.

```python
import numpy as np
```

```python
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

print(f"{'Month':<6} {'Mean':>7} {'Min':>7} {'Max':>7} {'Std':>7}")

print("-" * 38)

for i, names in enumerate(month_names):
    column = np.loadtxt("../csv/monthly_temperatures.csv", delimiter = ",", skiprows = 1, usecols = i)
    print(f"{names:<6} {np.mean(column):>7.1f} {np.min(column):>7.1f} {np.max(column):>7.1f} {np.std(column):>7.1f}")
```

## Saving to a csv file

```python
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
results = []

for i, names in enumerate(month_names):
    column = np.loadtxt("../csv/monthly_temperatures.csv", delimiter = ",", skiprows = 1, usecols = i)
    results.append([names, np.mean(column), np.min(column), np.max(column), np.std(column)])

results = np.array(results, dtype = object)
np.savetxt("../csv/monthly_stats.csv", results, delimiter = ",", fmt = ["%s", "%.1f", "%.1f", "%.1f", "%.1f"], header = "Month, mean, min, max, std", comments = "")
```
