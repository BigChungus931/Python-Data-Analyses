```python
import numpy as np
```

```python
jan = np.loadtxt("../csv/monthly_temperatures.csv", delimiter = ",", skiprows = 1, usecols = 0)

max_jan = np.max(jan)
min_jan = np.min(jan)
std_jan = np.std(jan)
mean_jan = np.mean(jan)

print(f"Max = {max_jan}")
print(f"Min = {min_jan}")
print(f"Std = {std_jan}")
print(f"Mean = {mean_jan}")
```
