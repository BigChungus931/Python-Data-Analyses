```python
import numpy as np
import random
```

```python
arr = np.arange(1, 11)
print(arr)
```

```python
arr = np.zeros((3, 3))
arr[1, 1] = 9

print(arr)
```

```python
arr = np.linspace(0, 1, 5)
print(arr)
```

```python
arr = np.arange(1, 13)
grid = arr.reshape(3, 4)
print(grid)
```

```python
arr = np.ones((4, 3))
print('Shape:', arr.shape)
print('Rows:', arr.shape[0])
print('Total elements:', arr.size)
```

```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])
print('Sum:', a + b)
print('Product:', np.dot(a, b))
```

```python
celsius = np.array([0, 20, 37, 100])
fahrenheit = celsius * 9/5 + 32
print(fahrenheit)
```

```python
arr = np.array([4, 8, 15, 16, 23, 42])
print('Sum:', np.sum(arr))
print('Mean:', np.mean(arr))
print('Min:', np.min(arr))
print('Max:', np.max(arr))
```

```python
arr = np.array([5, 10, 15, 20, 25, 30])
print('First:', arr[0])
print('Last:', arr[-1])
print('Middle slice', arr[2:5])
```

```python
arr = np.arange(1, 17).reshape(4, 4)
print('Second row:', arr[1])
print('Third column:', arr[:,2])
```

```python
dice_rolls = np.random.randint(1, 7, size=6)
print('Dice rolls:', dice_rolls)
```

```python
random_grid = np.random.rand(3, 3)
print('Grid:', random_grid)
print('Average:', np.mean(random_grid))
```

```python
arr = np.array([3, 7, 2, 9, 4, 11, 6])
big_numbers = arr[arr > 5]
print('Numbers > 5:', big_numbers)
```

```python
arr = np.array([42, 7, 19, 3, 88, 55, 1])
sorted_arr = np.sort(arr)
position = np.where(sorted_arr == 42)
print('Sorted:', sorted_arr)
print('42 is at index:', position)
```

```python
marks = np.array([72, 85, 60, 91, 55, 78, 88, 66])

average = np.mean(marks)

print('Average:', average)
print('Highest:', np.max(marks))
print('Lowest:', np.min(marks))

passed = marks[marks >= 60]
print('Student passed:', passed.size)

above_avg = marks[marks >= average]
print('Above average marks:', above_avg.size)
```
