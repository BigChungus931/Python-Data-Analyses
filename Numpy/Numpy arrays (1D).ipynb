# Numpy Arrays 1D
### What are Numpy Arrays
A Numpy Array is a grid of values all of the same type stored in a continuous block of memory. A 1D array is the simplest form - is just a single row of values like a list but far more powerful.

```python
import numpy as np #Import section
```

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr)

print(type(arr))
```

### Creating arrays

---
There are several ways to create a 1D array depending on what you need <br>
**from a Python list**

```python
a = ([1,2,3,4,5])
```

**Range of values** -np.arange() <br>

- Start is the starting value 
- End is the ending value
- Step is the difference between each values in the array


```python
arr1 = np.arange(5)
arr2 = np.arange(1, 11) #(start and end)
arr3 = np.arange(0, 11, 3) #(start, end, step)
arr4 = np.arange(10, 0, -1)
print(arr4)
```

**Evenly spaced values** <br>

To space an array evenly across numbers, for example if you are given an array from 1 to 100. <br>
When you use linspace, the array will be divided into 10s.

```python
arr = np.linspace(0, 100, 11)
arr2 = np.linspace(0,1, 10)

print(arr2)
```

**Zeros and ones**

```python
ones = np.ones(10)

zeros = np.zeros(10)

print(zeros)
```
