# Cleaning Data

```python
import pandas as pd
```

```python
df = pd.read_excel("Datasets/signup.xlsx")
df
```

<h2 style = "color:#4fd4ff" >Detecting Missing Values</h2>

Before you can fix missing data, you need to find it. <br>
Pandas represents a missing value as NaN (Not a Number). <br>
isnan() returns a same-shaped DataFrame of True/False. <br>
Returns True whenever a cell is missing.

```python
df.isna()
```

Knowing that missing values exist is step 1. <br>
Knowing exactly how many, and in which columns, is what turns that into a plan. <br>
Chaining .sum() onto .isna() returns the count of missing values in each column.

```python
df.isna().sum()
```

```python
(df.isna().sum()/len(df) * 100).sort_values(ascending = False) # Percentage missing per column
```

```python
df.isna().sum().sum() # Total missing cells in the entire DataFrame
```

```python
df.info() # shows non-null count per column which is a mirror image of this command df.isna().sum()
```

<h2 style = "color:#4fd4ff" >Dropping Rows or Columns - dropna()</h2>

```python
df.dropna() # Drops any row with at least one missing value
```

```python
df.dropna(axis = 1) # Drops any column with a missing value instead of rows
```

```python
df.dropna(how = "all") # Only drops a row if all its values are missing
```

```python
df.dropna(subset = ["name", "email"]) # Only considers missing values in specific columns
```

<h2 style = "color:#4fd4ff" >Duplicate Rows</h2>

Data collected from multiple sources - for example a form resumbited, 2 systems  merged, a scrapper run twice, almost always produces duplicate rows.<br>
.duplicated() flags them <br>
.drop_duplicates() removes them

```python
#df = pd.read_excel("Datasets/kids_records.xlsx")
#df
```

```python
df.isna().sum()
```

```python
(df.isna().sum()/len(df) * 100).sort_values(ascending = False)
```

```python
df = df.dropna(how = "all")
```

```python
df.dropna(subset = ["name"])
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
dup_emails = df[df.duplicated(subset = ["email"], keep = False)].sort_values("email")
print(dup_emails)
```

```python
students_df = pd.read_excel("Datasets/school_kids_100.xlsx")
students_df
```

```python
students_df = students_df.dropna(how = "all")
```

```python
students_df.duplicated().sum()

students_df = students_df.drop_duplicates()
```

```python
students_df.duplicated().sum()
```

```python
students_df = students_df.dropna(subset = ["student_id"])
students_df
```

```python
print(len(students_df))
```

```python
students_df["gender"] = students_df["gender"].str.strip().str.upper().map({
    "m" : "Male",
    "M" : "Male",
    "MALE" : "Male",
    "male" : "Male",
    "female" : "Female",
    "FEMALE" : "Female", 
    "f" : "Female",
    "F" : "Female"
})
```

```python
students_df
```

```python
students_df.to_excel("Datasets/students_clean.xlsx", index = False)
```
