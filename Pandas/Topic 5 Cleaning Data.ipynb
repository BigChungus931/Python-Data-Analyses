<h2 style = "color:#4fd4ff" >Detecting Missing Values</h2>

Before you can fix missing data, you need to find it. <br>
Pandas represents a missing value as NaN (Not a Number). <br>
isnan() returns a same-shaped DataFrame of True/False. <br>
Returns True whenever a cell is missing.

```python
df.isna()
```

Knowing that missing values exist is step 1. <br>
Knowing exactly how many, and in which columns, is what turns that into a plan. <br>
Chaining .sum() onto .isna() returns the count of missing values in each column.

```python
df.isna().sum()
```

```python
(df.isna().sum()/len(df) * 100).sort_values(ascending = False) # Percentage missing per column
```

```python
df.isna().sum().sum() # Total missing cells in the entire DataFrame
```

```python
df.info() # shows non-null count per column which is a mirror image of this command df.isna().sum()
```
