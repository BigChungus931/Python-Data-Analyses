<h2 style = "color:#4fd4ff" >What is Pandas and why do they matter?</h2>

1. Think about Microsoft Excel. You have rows, columns, headers, and data. You can filter, sort, calculate totals, and look things up. pandas does all of that but in Python code instead of clicking around a spreadsheet.

2. The big difference is that code is reproducible. If you clean a dataset in Excel and someone asks "how did you do that?", you have to explain every click. In pandas you just show them the code, ten lines that do the exact same thing every time.

```python
import pandas as pd
```

<h2 style = "color:#4fd4ff" >Two core structure : DataFrame and Series</h2> 

<h3 style = "color:#d4ff4f" >Series - One column of data</h3> 

A series is like a single column from a spreadsheet. <br>
it has values and index.

```python
salaries = pd.Series(
    [4000, 2500, 6000, 1000, 5700],
    index = ["John", "Mary", "Mike", "Mia", "Alex"] # Replacing the index numbers with names
)

print(salaries[:]) # printing the entire series
print(salaries["Mike"]) # Printing a specific salary using the index name rather than the index number
```

<h3 style = "color:#d4ff4f" >DataFrame - Full Table</h3> 

A DataFrame is a full table - multiple columns together. <br>
Think if it as a spreadsheet in Python. Each column in a DataFrame is a series. <br>
They are just sitting side by side sharing the same index.

```python
data = {
    "Name" : ["John", "Mary", "Mike", "Mia", "Alex"],
    "Salaries" : [4000, 2500, 6000, 1000, 5700],
    "Dept" : ["IT", "HR", "IT", "Finance", "IT"]
}

df = pd.DataFrame(data)

print(df)

dept_col = df["Dept"]

print(dept_col)

print(type(df))
```
