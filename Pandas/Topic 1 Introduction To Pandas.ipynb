## What are Pandas and why do they matter?

1. Think about Microsoft Excel. You have rows, columns, headers, and data. You can filter, sort, calculate totals, and look things up. pandas does all of that but in Python code instead of clicking around a spreadsheet.

2. The big difference is that code is reproducible. If you clean a dataset in Excel and someone asks "how did you do that?", you have to explain every click. In pandas you just show them the code, ten lines that do the exact same thing every time.
