# Selecting and Filtering Data

```python
import pandas as pd
```

<h2 style = "color:#4fd4ff" >Selecting Single and Multiple Columns</h2>

We already met bracket notation in topic 3. <br>
It is also our entry point into filtering, because the same square brackets that pull out the columns also pull out rows. <br>
The trick is knowing what to put inside them.

```python
df = pd.read_csv("Datasets/students-csv.csv")
print(df)
print(df["city"])
print(df[["names", "city"]])
```

<h2 style = "color:#4fd4ff" >Boolean Filtering - Selecting Rows by Conditions</h2>

```python
condition = (df["city"] == "Miami") | (df["city"] == "London")

print(df[condition])
```

```python
condition = (df["pass"] == True) & (df["gender"] == "Male") & (df["scores"] >= 95)
print(df[condition])
```

<h2 style = "color:#4fd4ff" >The .query() Method - Readable Filtering</h2>

.query() takes a plain string expression instead of bracket and mask syntax. <br>
For anyone who finds a wall of & and parentheses hard to read, it is a cleaner alternative and for very large DataFrames it can be faster.

```python
df.query('city == "London" or city == "Miami"')
```

```python
df.query('`pass` == True and gender == "Male" and scores >= 95')
```

<h2 style = "color:#4fd4ff" >.isin() - Filtering by a List of Values</h2>

Checking a column against several possible values with | gets repetitive fast. <br>
.isin() takes a list and matches any row whose value appears anywhere in it — one condition instead of many.

```python
df[df["city"].isin(["NYC", "Miami", "Amsterdam"])]
```

<h2 style = "color:#4fd4ff" >.between() - Filtering Within a Range</h2>

A range filter — value above a floor and below a ceiling — normally takes two chained conditions. <br>
.between() does it in one call, and includes both endpoints by default.

```python
df[df["scores"].between(88, 98)] # The start and the stop is included
```

```python
df[df["scores"].between(88, 98, inclusive = "neither")] # The start and the stop is not included
```

<h2 style = "color:#4fd4ff" >Selecting a Random Sample Using .sample()</h2>

Not every selection is about a condition. <br>
Sometimes you just want a quick, honest look at scattered rows — for spot-checking data quality, or splitting off a test set for a model. <br>
.sample() grabs a random subset.

```python
df.sample(3, random_state = 10)
```
