# Pandas Series

<h2 style = "color:#4fd4ff" >Creating a Series</h2>

A Series is a labeled one-dimensional array. Think of it as a single column of data where every row has both a value and a label. You can create a Series from three common sources: a plain Python list, a dictionary, or a NumPy array.

```python
import pandas as pd
import numpy as np
```

<h3 style = "color:#d4ff4f" >From a List</h3> 

This is the simplest way of making a series. All you need to do is pass a list into the pd.Series <br> 
method and pandas automatically assigns integer labels (index) starting from 0.

```python
scores = pd.Series([100, 400, 300, 600, 500], index = ["Alica", "Bob", "Diana", "Max", "Eve"])

print(scores)
```

<h3 style = "color:#d4ff4f" >From a Dictionary</h3> 

When you pass a dictionary, the keys become the index labels and the values become the data.

```python
scores = pd.Series({
    "Alica" : 88,
    "Bob" : 95,
    "Diana" : 61,
    "Max" : 72,
    "Eve" :  55
})

print(scores)
```

<h3 style = "color:#d4ff4f" >From a Numpy Array</h3> 

Since Panda is built on top of Numpy, you can pass a Numpy array directly. <br>
The result is identical to passing a list.

```python
arr = np.array([68, 32, 57, 10, 22])

scores = pd.Series(arr)

print(scores)
```

<h3 style = "color:#bf9fff" >Practice</h3>
