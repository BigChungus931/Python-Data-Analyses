# Pandas DataFrames

<h2 style = "color:#4fd4ff" >Creating a DataFrame from a dictionary</h2>

A DataFrame is made up of multiple Series. <br>
A DataFrame is a 2 dimensional rows and columns. <br>
The most direct way to build one from scratch is to pass a dictionary to **pd.DataFrame()**.

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "names" : ["Alex", "Bob", "Diana", "Jane"],
    "scores" : [88, 95, 72, 61],
    "pass" : [True, True, True, False],
    "city" : ["NYC", "London", "Naples", "Seattle"]
}

df = pd.DataFrame(data)

print(df)
```

<h2 style = "color:#4fd4ff" >Reading Files</h2>

In practice, you rarely build DataFrames by hand. <br>
You load them from files. <br>
The 3 most common formats are csv, excel, and JSON.

<h3 style = "color:#d4ff4f" >read_csv</h3> 

pd.read_csv() is the most used function in pandas.<br>
Pass a file path through it and it returns a DataFrame.

```python
df = pd.read_csv("Datasets/students-csv.csv")

df2 = pd.read_csv(
    "Datasets/students-csv.csv",
    usecols = ["names", "scores"],
    nrows = 100,  
)

print(df)
print(df2)
```

<h3 style = "color:#d4ff4f" >read_excel and read_json</h3> 

```python
df = pd.read_excel("Datasets/students-excel.xlsx", sheet_name = "students")

print(df)
```

```python
df = pd.read_json("Datasets/student data.json")

print(df)
```

<h2 style = "color:#4fd4ff" >Writing Files</h2>

<h3 style = "color:#d4ff4f" >to_excel and to_csv</h3> 

Once you have cleaned or transformed your data, you will want to save it. <br>
Every read function has a matching write function.

```python
df = pd.read_csv("Datasets/students-csv.csv")
print(df.head()) # Previews the first five rows
print(df.tail()) # Previews the last five rows
print(df.shape) # Number of rows and columns
print(df.info()) # Information on the file
print(df.describe()) # Info below
print(df.columns) # List of all the columns
```

**.describe()** runs summary statistics on every numeric column at once. <br>
count, mean, std, min, the 3 quartiles, max.

<h5>Cleaning and Transforming Data</h5>

Common cleaning steps include handling missing values, fixing datatypes, renaming columns, and filtering rows.

```python
df = df.dropna() # Drops rows with any missing values
df = df.rename(columns = {"scores":"grades"}) # Changes a specific column
df = df[df["grades"] > 60] # Filters out data from the file
df = df.drop_duplicates() # Removes duplicate values
df = df.to_csv("Datasets/cleaned.csv", index = False)
print(df)
```
