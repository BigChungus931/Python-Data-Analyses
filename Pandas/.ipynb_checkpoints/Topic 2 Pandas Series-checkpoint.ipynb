# Pandas Series

<h2 style = "color:#4fd4ff" >Creating a Series</h2>

A Series is a labeled one-dimensional array. Think of it as a single column of data where every row has both a value and a label. You can create a Series from three common sources: a plain Python list, a dictionary, or a NumPy array.

```python
import pandas as pd
import numpy as np
```

<h3 style = "color:#d4ff4f" >From a List</h3> 

This is the simplest way of making a series. All you need to do is pass a list into the pd.Series <br> 
method and pandas automatically assigns integer labels (index) starting from 0.

```python
scores = pd.Series([100, 400, 300, 600, 500], index = ["Alica", "Bob", "Diana", "Max", "Eve"])

print(scores)
```

<h3 style = "color:#d4ff4f" >From a Dictionary</h3> 

When you pass a dictionary, the keys become the index labels and the values become the data.

```python
scores = pd.Series({
    "Alica" : 88,
    "Bob" : 95,
    "Diana" : 61,
    "Max" : 72,
    "Eve" :  55
})

print(scores)
```

<h3 style = "color:#d4ff4f" >From a Numpy Array</h3> 

Since Panda is built on top of Numpy, you can pass a Numpy array directly. <br>
The result is identical to passing a list.

```python
arr = np.array([68, 32, 57, 10, 22])

scores = pd.Series(arr)

print(scores)
```

<h2 style = "color:#4fd4ff" >Series Index - Label System</h2>

Every value in a series has a label, that label is called the **index**. <br>

By default it's just 0, 1, 2,.... <br>

But we can set the index to anything like str, dates, custom numbers.

```python
temperatures = pd.Series([29, 31, 27, 22, 35, 18, 41], index = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])

temperatures.index.name = "Days"

temperatures.name = "Temp Celsius"

temperatures.index = "M", "T", "W", "Th", "F", "Sa", "Su"

print(temperatures)
```

The index is not just decoration. It is the address system pandas uses to align data when you combine Series, do arithmetic between them, or merge DataFrames.

```python
a = pd.Series([10, 20, 30], index = ["x", "y", "z"])

b = pd.Series([1, 2, 3], index = ["y", "z", "x"])

print(a + b)
```

<h2 style = "color:#4fd4ff" >Accessing Values - .iloc vs .loc</h2>

This is where beginners make the most mistakes. pandas gives you two ways to access data in a Series, and they work completely differently.

```python
scores = pd.Series(
    [88, 95, 72, 61, 90],
    index = ["Alice", "Bob", "Carlos", "Diana", "Eve"]
)

# iloc and loc
print(scores.iloc[1:4])
print(scores.loc["Bob":"Diana"])
```

**.iloc** stands for integer location, it uses the integer position of the value like indexing a list. <br>
**.loc** uses the index label if your label are strings.

With .iloc slicing, the end index is excluded (like Python lists). With .loc slicing, the end label is included. <br>
This inconsistency trips up everyone at first — now you know.

<h2 style = "color:#4fd4ff" >Vectorized operations on a series</h2>

One of the biggest advantages of a Series is that you can apply an operation to every value at once <br> — no loops needed. This is called **vectorization**, and it is inherited from NumPy.

<h3 style = "color:#d4ff4f" >Math Operations</h3> 

Arithmetic operators apply element-wise across the entire series.

```python
salaries = pd.Series([4000, 2500, 8900, 6300], index=["John", "Mary", "Mike", "Victor"])

raised = salaries * 1.1

print(raised)

above_avg = salaries - salaries.mean()

print(above_avg)
```

<h3 style = "color:#d4ff4f" >Boolean Conditions</h3> 

You can use boolean conditions to filter the original series.

```python
filters = salaries > 4000

print(salaries[filters])
```

<h3 style = "color:#d4ff4f" >String Operations Via .str</h3> 

For text data, pandas provide a .str accessor that apply string methods to every element at once.

```python
names = pd.Series(["alex", "richard", "dan", "will"])

print(names.str.upper())
print(names.str.capitalize())
print(names.str.len())
print(names[names.str.startswith("a")])
print(names[names.str.contains("ar")])
```

<h3 style = "color:#d4ff4f" >Applying Custom Functions</h3> 

When built-in operations are not enough, use .apply() to run any python function on every element.

```python
salaries = pd.Series([4000, 2500, 8900, 6300], index=["John", "Mary", "Mike", "Victor"])

def tax_brackets(salaries):
    if salaries >= 6300:
        return "High"

    elif salaries >= 3000:
        return "Mid"

    else:
        return "Low"

brackets = salaries.apply(tax_brackets)

print(brackets)
```
