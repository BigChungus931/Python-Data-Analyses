# Pandas DataFrames

<h2 style = "color:#4fd4ff" >Creating a DataFrame from a dictionary</h2>

A DataFrame is made up of multiple Series. <br>
A DataFrame is a 2 dimensional rows and columns. <br>
The most direct way to build one from scratch is to pass a dictionary to **pd.DataFrame()**.

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "names" : ["Alex", "Bob", "Diana", "Jane"],
    "scores" : [88, 95, 72, 61],
    "pass" : [True, True, True, False],
    "city" : ["NYC", "London", "Naples", "Seattle"]
}

df = pd.DataFrame(data)

print(df)
```

<h2 style = "color:#4fd4ff" >Reading and Writing Files</h2>

In practice, you rarely build DataFrames by hand. <br>
You load them from files. <br>
The 3 most common formats are csv, excel, and JSON.

<h3 style = "color:#d4ff4f" >read_csv</h3> 

pd.read_csv() is the most used function in pandas.<br>
Pass a file path through it and it returns a DataFrame.

```python
df = pd.read_csv("Datasets/students-csv.csv")

df2 = pd.read_csv(
    "Datasets/students-csv.csv",
    usecols = ["names", "scores"],
    nrows = 100,  
)

print(df)
print(df2)
```

<h3 style = "color:#d4ff4f" >read_excel and read_json</h3> 

```python
df = pd.read_excel("Datasets/students-excel.xlsx", sheet_name = "students")

print(df)
```

```python
df = pd.read_json("Datasets/student data.json")

print(df)
```
