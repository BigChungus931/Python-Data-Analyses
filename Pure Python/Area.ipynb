# Area of a right triangle
## Calculating the area of a right triangle
### Formula
$A = \frac{bh}{2}$ <br>
Calculate the area of a right triangle with base as 10 and height as 3


````python
b = 10
h = 3
A = (b*h)/2
print(A)
````
