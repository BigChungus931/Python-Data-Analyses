# Matplot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

### Line Graph

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

x_smooth = np.linspace(min(x), max(x), 300)
spline = make_interp_spline(x, y)
y_smooth = spline(x_smooth)
plt.plot(x_smooth, y_smooth)
plt.scatter(x, y, color = "red")
plt.title("Graph of x and y")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([10, 15, 20, 25, 30])
plt.grid(True)
```

Graph the following math problem

```python
x = np.linspace(-2, 2, 5)

y = 2*x + 3

plt.figure(figsize = (8, 6))
plt.plot(x, y, color = "blue", label = "y = 2x + 3")
plt.scatter(x, y, color = "red")
plt.title("Graph of y = 2x + 3")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid(True)
plt.yticks([-1, 0, 1, 2, 3, 4, 5, 6, 7])
plt.xticks([-2, -1, 0, 1, 2])
```

```python
x = np.linspace(-3, 3, 7)
y = x**2

x_smooth = np.linspace(min(x), max(x), 300)
spline = make_interp_spline(x, y)
y_smooth = spline(x_smooth)
plt.figure(figsize = (10, 3))
plt.plot(x_smooth, y_smooth, color = "blue", label = "y = x²")
plt.scatter(x, y, color = "red")
plt.title("Graph of y = x²")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid(True)
```
